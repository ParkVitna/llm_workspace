## Fine-tuned 모델 추론

```python
from google.colab import userdata
from openai import OpenAI
import os

# API_KEY 를 명시적으로 전달하는 방법 - 이걸 더 추천!
OPENAI_API_KEY = userdata.get('MY_OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

# 환경변수에서 API_KEY 를 전달하는 방법
# os.environ("OPENAI_API_KEY") = userdata.get("OPENAI_API_KEY")
# client = OpenAI()
```

```python
model = "ft:gpt-4o-mini-2024-07-18:personal:sarcastic-chat-bot:Bms1g0D0"

def sarcastic_bot(prompt, temperature=1.0):
  stream = client.chat.completions.create(
      model=model,
      messages=[
          {"role":"system", "content":"너는 사실을 말하는 챗봇이지만, 빈정대거나 비꼬는 말투로 응답하는 고장난 챗봇이다."},
          {"role":"user"  , "content":prompt},
      ],
      temperature=temperature,
      stream=True
  )
  for chunk in stream:
    content = chunk.choices[0].delta.content
    if content:
      print(content, end="")
```

```python
sarcastic_bot("외국인 여자친구를 만들려면 어떡해야되?")
```
