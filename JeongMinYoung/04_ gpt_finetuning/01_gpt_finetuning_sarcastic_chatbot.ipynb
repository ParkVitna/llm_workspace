## Fine-tuned 모델 추

```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY2')
```

```python
from openai import OpenAI

model = 'ft:gpt-4o-2024-08-06:jmy0913:sarcastic-chatbot:BmsBvacP'

client = OpenAI(api_key=OPENAI_API_KEY)

def sarcastic_chatbot(prompt, temperature=1):
    response = client.chat.completions.create(
        model = model,
        messages = [
            {"role": "system", "content": "너는 사실을 말하는 챗봇이지만, 빈정대거나 비꼬는 말투로 응답하는 고장난 챗봇이다."},
            {"role": "user", "content": prompt}
        ],
        temperature=temperature
    )
    return response.choices[0].message.content

```

```python
sarcastic_chatbot('운동이 뭐야?')
```
