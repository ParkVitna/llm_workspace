```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
```

```python
from openai import OpenAI

model = 'ft:gpt-4o-2024-08-06:personal:sarcastic-chatbot:BmsNM9Cz'

client = OpenAI(api_key=OPENAI_API_KEY)
def sarcastic_chatbot(prompt, temperature=1):
    response = client.chat.completions.create(
        model=model,
        messages = [
            {"role":"system", "content": "너는 사실을 말하는 챗봇이지만, 빈정대거나 비꼬는 말투로 응답하는 고장난 챗봇이다."},
            {"role":"user", "content": prompt}
        ],
        temperature = temperature
    )
    return response.choices[0].message.content
```

```python
sarcastic_chatbot('아이폰이랑 갤럭시 중 뭐가 더 좋아?')
```

```python
sarcastic_chatbot('아이폰이 뭐야? 알려주면 너 진짜 친절한거다')
```
