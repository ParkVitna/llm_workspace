## 데이터준비

```python
import json

with open('KorQuAD_v1.0_dev.json','r', encoding='utf-8') as f:
    dev_data = json.load(f)

dev_data['data'][0]
```

```python
# 데이터 정제

items = [item for topic in dev_data['data'] for item in topic['paragraphs']]
qa_dict = {}

for item in items:
    for entry in item['qas']:
        question = entry['question']
        answer = [ans['text'] for ans in entry['answers']][0]
        qa_dict[question] = answer

len(qa_dict)

# qa_dict = {}

# for item in items:
#     for entry in item['qas']:
#         question = entry['question']
#         answer = [ans['text'] for ans in entry['answers']][0]
#         qa_dict[question] = answer

# len(qa_dict)
```

```python
qa_dict = dict(list(qa_dict.items())[:20])
qa_dict
```

```python
results = []

for q, a in qa_dict.items():
    qa = {
        "messages": [
            {"role": "system", "content": ""},
            {"role": "user", "content": q},
            {"role": "assistant", "content": a},
        ]
    }
    results.append(qa)

results
```

```python
# jsonl 파일 생성
with open('korquad_data.jsonl', 'w', encoding='utf-8') as f:
    for qa in results:
        json_str = json.dumps(qa, ensure_ascii=False)
        f.write(json_str + '\n')
```

## Fine-tuned 모델 추론

```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
```

```python
from openai import OpenAI

model = ''

client = OpenAI(api_key=OPENAI_API_KEY)

def sarcastic_chatbot(prompt, temparature=1):
    response = client.chat.completions.create(
        model = model,
        message = [
            {"role": "system", "content" : "너는 사실을 말하는 챗봇이지만, 빈정대거나 비꼬는 말투로 응답하는 고장난 챗봇이다."},
            {"role": "user", "content":prompt}
        ],
        temperature = temparature
    )
    return response.choices[0].message.content
```

```python
from openai import OpenAI

model = 'ft:gpt-4.1-2025-04-14:person:korquad20:BmsvOCUK'

client = OpenAI(api_key=OPENAI_API_KEY)

def korquad_finetuned_qa(prompt, temparature=1):
    response = client.chat.completions.create(
        model = model,
        messages = [
            {"role": "system", "content" : ""},
            {"role": "user", "content":prompt}
        ],
        temperature = temparature
    )
    return response.choices[0].message.content
```

```python
korquad_finetuned_qa('알렉산더 헤이그와 1950년 5월 결혼한 상대의 이름은 무엇인가?')
```
