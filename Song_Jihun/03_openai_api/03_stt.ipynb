# STT Speech To Text
OpenAI 에 STT 라는 이름으로는 없고, OpenAI 에서 제공하는거 이름은 `Whisper`.

## Whisper

[Whisper 옛날 링크 (https://platform.openai.com/docs/models#whisper)](https://platform.openai.com/docs/models#whisper) <br/>
[Whisper 최신 링크 (https://platform.openai.com/docs/models/whisper-1)](https://platform.openai.com/docs/models/whisper-1)

Whisper는 OpenAI에서 개발한 범용 음성 인식 모델로, 다양한 오디오 데이터셋을 학습하여 다국어 음성 인식, 음성 번역, 언어 식별 등의 작업을 수행할 수 있다.

Whisper v2-large 모델은 현재 API를 통해 'whisper-1'이라는 이름으로 제공되고 있다.

오픈 소스 버전의 Whisper와 API를 통한 Whisper는 기능적으로 동일하지만, API를 통해 제공되는 버전은 최적화된 추론 과정을 거쳐 다른 방법에 비해 더 빠르게 동작한다.

```python
from google.colab import userdata
from openai import OpenAI
import os

# API_KEY 를 명시적으로 전달하는 방법 - 이걸 더 추천!
OPENAI_API_KEY = userdata.get('MY_OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

# 환경변수에서 API_KEY 를 전달하는 방법
# os.environ("OPENAI_API_KEY") = userdata.get("OPENAI_API_KEY")
# client = OpenAI()
```

```python
file_path = "output.mp3"
with open(file_path, "rb") as f:
  transcription = client.audio.transcriptions.create(
      model="whisper-1",
      file=f
  )
print(transcription)
print(transcription.text)
```
